Задание №1

Работа с библиотекой numpy. Создайте вектор целых чисел от 0 до 9
включительно длиной 20 чисел. Найти наиболее часто встречающееся
значение в массиве и выведите его.

```python
#Задание №1

import numpy as np

arr = np.random.randint(0,9,20)
unique, counts = np.unique(arr, return_counts=True)
print('Вектор целых чисел: ',arr)
print('Наиболее часто встречающееся значение:',unique[np.argmax(counts)])
```

Задача №2

  Работа с библиотекой Pandas. Загрузить файл ‘Ecommerce Purchases'  - электронные покупки.

  А) Сколько человек имеют должность «Lawyer»
  
  Б) Сколько человек имеют American Express в качестве поставщика кредитных карт (CC Provider) и совершили покупку выше $95 (Purchase Price)

```python
#Задача №2

import pandas as pd

data = pd.read_csv('Ecommerce Purchases.csv')

print('Кол-во сотрудников с должностью Lawyer :',data['Job'].tolist().count('Lawyer'))

df = data[['CC Provider','Purchase Price']]

df = df[df['Purchase Price'] > 95]

print('Кол-во человек имеющих American Express и >95$:', len(df[df['CC Provider'] == 'American Express'].axes[0]))
```

Задача №3

 A) проходящий через все точки

 B) график двух функций: первая парабола через первые 3 точки,вторая парабола через следующие 3 точки

```python
from re import X
import matplotlib.pyplot as plt

import numpy as np

x = np.random.randint(0,100,6)
x.sort()
y = x ** 2 + 5 * x + 2
plt.scatter(x, y, c ="blue")
plt.plot(x,y)
plt.show()
x1 = np.random.randint(x[0:2].min(),x[0:3].max(),100)
x2 = np.random.randint(x[3:6].min(),x[3:6].max(),100)

x1.sort()
x2.sort()

y1 = x1 ** 2 + 2 * x1 - 3

y2 = -x2 ** 2 +2 * x2 + 5
plt.scatter(x, y, c ="blue")

plt.plot(x1, y1,'b--', x2, y2, 'g' )
```
